# Task 1 - Plot a line graph to show the difference between heart rate of smokers and non smokers

```python
#Import libraries

#read the csv
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('framingham.csv')
df
```

```python
#Filter and make a new dataframe for non smokers
non_smokers = df.loc[df['currentSmoker'] == 0]
non_smokers
```

```python
#Group by age column and find average heart rate at different age
group_non_smokers = non_smokers.groupby('age')['heartRate'].mean().reset_index()
group_non_smokers
```

```python
#Filter and make a new dataframe for smokers
smokers = df.loc[df['currentSmoker'] == 1]
smokers
```

```python
#Group by age column and find average heart rate at different age
group_smokers = smokers.groupby('age')['heartRate'].mean().reset_index()
group_smokers
```

```python
#Plot a line graph to show the heart rate of smokers vs non smokers
label = group_non_smokers['age']
value = group_non_smokers['heartRate']

fig = plt.subplots(figsize=(19,8))

plt.plot(label, value,label="Heart rate of non smokers in various age groups",linewidth=3.0)

label = group_smokers['age']
value = group_smokers['heartRate']
plt.plot(label, value,label="Heart rate of smokers in various age groups",linewidth=3.0)

plt.xlabel('Ages')
plt.xticks(rotation='vertical')

plt.ylabel('Heart Rates')

plt.title('The different heart rates between smokers and non smokers in various age groups', fontsize=20)

plt.legend()

plt.show()
```

Conslusion - When you don't smoke your heart rate is more stable then when smoking.

# Task 2 - Which age group have high chances of having coronary heart disease in 10 years

```python
#Read the csv
df1 = pd.read_csv('framingham.csv')
df1
#Filter and make a new dataframe for those who has chances of having coronary heart disease in 10 years
coronary_heart_disease=df1.loc[df1['TenYearCHD'] == 1]
coronary_heart_disease
```

```python
#Group by age column and count the rows of TenYearCHD column
group_age = coronary_heart_disease.groupby('age')['TenYearCHD'].count().reset_index()
group_age
```

```python
#Plot a line graph to show total number of people having a chance of coronary heart disease in 10 years
plt.figure(figsize=(20, 8))

plt.scatter(group_age['age'], group_age['TenYearCHD'], color = "blue", label='Total number of people having a chance of coronary heart disease in 10 years', alpha = 0.5, s= group_age['TenYearCHD']*5)

plt.xticks(rotation='vertical')
plt.legend()
plt.xlabel("Age", size=14)
plt.ylabel("Chance of coronary heart disease in 10 years", size=14)
plt.title("total number of people having a chance of coronary heart disease in 10 years")
```

Conslusion - As your age starts going up the chances go up but at a mid point the chances start to drop again.
